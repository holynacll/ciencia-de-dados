```python
for i in df.select_dtypes(exclude='object').columns:
    print('______________________________________________________')
    print(i)
    print(df.groupby(['Survived'])[i].describe())
```

```python
# SEX
women = df.loc[df.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

men = df.loc[df.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print(rate_women, rate_men)
```

```python
# Pclass
pc1 = df.loc[df.Pclass == 1]["Survived"]
rate_pc1 = sum(pc1)/len(pc1)

pc2 = df.loc[df.Pclass == 2]["Survived"]
rate_pc2 = sum(pc2)/len(pc2)

pc3 = df.loc[df.Pclass == 3]["Survived"]
rate_pc3 = sum(pc3)/len(pc3)

print(rate_pc1, rate_pc2, rate_pc3)
```

```python
cat_feats = ['Sex']
cat_feats
```
